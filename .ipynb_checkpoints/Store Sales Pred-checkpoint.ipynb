## handling null values

- point to be noted that only outlets of supermarket_type1 and grocery type stores are having null values

```python
# First of all, we create a duplicate copy of the real dataframe for saftey measures
df = real_df.copy()

# item_weight is depending upon item_type
# So, we replacing null values simply by the corresponding mean value of the item_type

# To gather mean values of various item_type, we create a dictionary

d1 = df.groupby('Item_Type').mean()
d1 = dict(zip(d1.index.to_list(), d1['Item_Weight'].to_list()))

# To get the indices of null values in item_weight column, we create a null_df
null_df = df[df['Item_Weight'].isna()==True]

# Now, we create a ordered list of mean values for the item_weight column by using the dictionary
itm_wt = []
for typ in null_df['Item_Type']:
    itm_wt += [d1[typ]]

# Finally, we assign mean values to the corresponding null values by using indices of null_df
c=0
for i in null_df.index:
    df.loc[i, 'Item_Weight'] = itm_wt[c]
    c+=1

display(df.head(15))

# To verify successful null-removal from the df
#df['Item_Weight'].isna().to_list().count(True)
```
